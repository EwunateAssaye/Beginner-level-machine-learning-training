# Handling the missing values

### fillna

fillna with 0

```python
df1=df1.fillna(0)
df1
```

fillna with different user assigned values 

```python
df1=df1.fillna({
    'experience':'three',
    'test_score(out of 10)':5,
    'interview_score(out of 10)':5,
    'salary($)':3000
})
df1
```

```python
df1=df1.fillna(method='ffill')
df1
```

```python
df1=df1.fillna(method='ffill', axis='columns')
df1
```

fill with bfill (backward fill)

```python
df1=df1.fillna(method='bfill')
df1
```

```python
df1=df1.fillna(method='bfill', axis='columns')
df1
```
